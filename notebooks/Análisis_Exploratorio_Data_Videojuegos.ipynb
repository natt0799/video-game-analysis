```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calificaciones promedio
rating_mean = df['rating'].mean()
print(f"Calificación promedio de los juegos: {rating_mean:.2f}")

# Distribución de las calificaciones
plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=20, kde=True, color='blue')
plt.title("Distribución de Calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Filtrar juegos con calificaciones extremas (altas y bajas)
# Definir umbrales para calificaciones altas y bajas
high_rating_threshold = 4.5  # Calificación alta
low_rating_threshold = 2.0   # Calificación baja

# Filtrar juegos con calificaciones altas
high_rated_games = df[df['rating'] >= high_rating_threshold][['name', 'rating', 'reviews_count', 'playtime', 'metacritic', 'suggestions_count']].sort_values(by='rating', ascending=False)

# Filtrar juegos con calificaciones bajas
low_rated_games = df[df['rating'] <= low_rating_threshold][['name', 'rating', 'reviews_count', 'playtime', 'metacritic', 'suggestions_count']].sort_values(by='rating', ascending=True)

# Mostrar los primeros 5 juegos con calificaciones más altas y más bajas
print("🎯 Juegos con Calificaciones Más Altas:")
print(high_rated_games.head(5))

print("\n⚠️ Juegos con Calificaciones Más Bajas:")
print(low_rated_games.head(5))

# Visualizar juegos con calificaciones extremas
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Gráfico para juegos con calificación alta
sns.barplot(data=high_rated_games.head(5), x='name', y='rating', ax=axes[0], palette='viridis')
axes[0].set_title('Top 5 Juegos con Calificaciones Más Altas')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')
axes[0].set_xlabel('Nombre del Juego')
axes[0].set_ylabel('Calificación')

# Gráfico para juegos con calificación baja
sns.barplot(data=low_rated_games.head(5), x='name', y='rating', ax=axes[1], palette='magma')
axes[1].set_title('Top 5 Juegos con Calificaciones Más Bajas')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')
axes[1].set_xlabel('Nombre del Juego')
axes[1].set_ylabel('Calificación')

plt.tight_layout()
plt.show()
```

```python
# Revisar valores nulos en la columna 'rating'
print(df['rating'].isnull().sum())

# Ajustar el umbral para capturar más juegos
low_rating_threshold = 2.5
low_rated_games = df[df['rating'] <= low_rating_threshold][['name', 'rating', 'reviews_count', 'playtime', 'metacritic', 'suggestions_count']].sort_values(by='rating', ascending=True)
print(low_rated_games.head(5))
```

```python
# Análisis de Juegos con Calificaciones Más Altas
high_rated_games_analysis = high_rated_games.head(5)

# Mostrar estadísticas descriptivas de los juegos mejor calificados
print("📊 Estadísticas de los Juegos con Calificaciones Más Altas:")
print(high_rated_games_analysis.describe())

# Visualizar los factores clave
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Reseñas
sns.barplot(data=high_rated_games_analysis, x='name', y='reviews_count', ax=axes[0])
axes[0].set_title('Cantidad de Reseñas - Juegos Mejor Calificados')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')

# Playtime
sns.barplot(data=high_rated_games_analysis, x='name', y='playtime', ax=axes[1])
axes[1].set_title('Horas de Juego - Juegos Mejor Calificados')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

# Puntaje Metacritic
sns.barplot(data=high_rated_games_analysis, x='name', y='metacritic', ax=axes[2])
axes[2].set_title('Metacritic - Juegos Mejor Calificados')
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Relación entre rating y playtime
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='playtime', y='rating', alpha=0.6)
plt.title("Relación entre Horas de Juego y Calificaciones")
plt.xlabel("Horas de Juego (playtime)")
plt.ylabel("Calificación (rating)")
plt.show()

# Relación entre rating y metacritic
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='metacritic', y='rating', alpha=0.6, color='purple')
plt.title("Relación entre Metacritic y Calificaciones")
plt.xlabel("Puntaje Metacritic")
plt.ylabel("Calificación (rating)")
plt.show()

# Relación entre rating y reviews_count
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='reviews_count', y='rating', alpha=0.6, color='orange')
plt.title("Relación entre Cantidad de Reseñas y Calificaciones")
plt.xlabel("Cantidad de Reseñas (reviews_count)")
plt.ylabel("Calificación (rating)")
plt.show()
```

```python
# Crear categorías de tiempo de juego
df['playtime_category'] = pd.cut(
    df['playtime'],
    bins=[0, 50, 200, float('inf')],
    labels=['0-50 horas', '50-200 horas', '200+ horas']
)

# Calcular estadísticas por categoría
playtime_summary = df.groupby('playtime_category').agg({
    'rating': 'mean',
    'reviews_count': 'mean',
    'metacritic': 'mean'
}).reset_index()

print("📊 Resumen por Categoría de Tiempo de Juego:")
print(playtime_summary)

# Visualizar resultados
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.barplot(data=playtime_summary, x='playtime_category', y='rating', ax=axes[0])
axes[0].set_title('Calificación Promedio por Categoría de Tiempo de Juego')
axes[0].set_xlabel('Tiempo de Juego')
axes[0].set_ylabel('Calificación Promedio')

sns.barplot(data=playtime_summary, x='playtime_category', y='reviews_count', ax=axes[1])
axes[1].set_title('Reseñas Promedio por Categoría de Tiempo de Juego')
axes[1].set_xlabel('Tiempo de Juego')
axes[1].set_ylabel('Cantidad de Reseñas')

sns.barplot(data=playtime_summary, x='playtime_category', y='metacritic', ax=axes[2])
axes[2].set_title('Metacritic Promedio por Categoría de Tiempo de Juego')
axes[2].set_xlabel('Tiempo de Juego')
axes[2].set_ylabel('Puntaje Metacritic')

plt.tight_layout()
plt.show()
```

```python
# Calcular correlación entre playtime y rating
correlation_playtime_rating = df['playtime'].corr(df['rating'])
print(f"🔗 Correlación entre Playtime y Rating: {correlation_playtime_rating:.2f}")
```

```python
# Crear un gráfico de distribución de calificaciones por categoría de tiempo de juego
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df, x='playtime_category', y='rating')
plt.title('Distribución de Calificaciones por Categoría de Tiempo de Juego')
plt.xlabel('Tiempo de Juego')
plt.ylabel('Calificación')
plt.show()
```

```python
# Crear una nueva columna para décadas
df['release_year'] = df['released'].dt.year
df['release_decade'] = (df['release_year'] // 10) * 10
# Calcular promedio de calificaciones por década
decade_ratings = df.groupby('release_decade')['rating'].mean().reset_index()

# Visualizar
plt.figure(figsize=(12, 6))
sns.lineplot(data=decade_ratings, x='release_decade', y='rating', marker='o')
plt.title('Calificación Promedio por Década')
plt.xlabel('Década')
plt.ylabel('Calificación Promedio')
plt.grid(True)
plt.show()
```

```python
# Identificar años con calificaciones más altas y más bajas
top_years = df.groupby('release_year')['rating'].mean().sort_values(ascending=False).head(3)
low_years = df.groupby('release_year')['rating'].mean().sort_values().head(3)

print("🏆 Años con Calificación más Alta:")
print(top_years)
print("\n⚠️ Años con Calificación más Baja:")
print(low_years)
```

```python
# Extraer el año de la columna 'released' y crear una nueva columna 'release_year'
df['release_year'] = pd.to_datetime(df['released']).dt.year

# Agrupar por año de lanzamiento y calcular el promedio de calificación
yearly_ratings = df.groupby('release_year')['rating'].mean().reset_index()
```

```python
# Línea de tiempo con hitos clave
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_ratings, x='release_year', y='rating', marker='o')
plt.title('Calificaciones Promedio a lo Largo de los Años con Eventos Clave')
plt.xlabel('Año')
plt.ylabel('Calificación Promedio')

# Anotaciones de eventos clave
important_events = {
    2006: 'Lanzamiento PS3 y Xbox 360',
    2013: 'Lanzamiento PS4 y Xbox One',
    2020: 'Lanzamiento PS5 y Xbox Series X'
}

for year, event in important_events.items():
    plt.axvline(x=year, color='red', linestyle='--')
    plt.text(year, yearly_ratings['rating'].mean(), event, rotation=90, verticalalignment='bottom')

plt.grid(True)
plt.show()
```

```python
# Relación entre sugerencias y calificación
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='suggestions_count', y='rating', alpha=0.6, color='teal')
plt.title("Relación entre Sugerencias y Calificaciones")
plt.xlabel("Cantidad de Sugerencias (suggestions_count)")
plt.ylabel("Calificación (rating)")
plt.show()

# Correlación entre reviews_count, suggestions_count y rating
correlation_matrix = df[['rating', 'reviews_count', 'suggestions_count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlación entre Calificación, Reseñas y Sugerencias")
plt.show()
```
